### Observations on data type and missing values.
From the data above, there are eleven columns in the dataset, each of which contains detailed information about different video games. It is evident from the first examination that the data types in certain columns do not match their descriptions.

We'll construct a new binary column called score_is_tbd to make sure this distinction is maintained for analysis in the future. Actual scores or NaN values will be displayed as 0 in this column, whereas "tbd" will be written as 1 to indicate missing or to-be-determined scores.  

3. In order to provide a more contextualized method of handling missing data, `critic_score` will group missing values by `genre` and `rating` and substitute the median for those groups. # Calculating Total Sales for Each Platform

***The following is a computation of total sales for each platform and finding which platform has the greatest total sales.

### Display of Charts to Show Sales Made by Platforms Above.


```python
# Grouping data by platform & summing total sales
platform_sales = games.groupby('platform')['total_sales'].sum()

# Filtering out only platforms with over 50 million in sales & creating a list
platforms_over_50 = platform_sales[platform_sales > 50].index.tolist()

# Filtering out original df to only include platforms that sold over 50 mil
games_filtered = games[games['platform'].isin(platforms_over_50)] #* Base DF

# Group the filtered DataFrame by 'platform'
grouped_by_platform = games_filtered.groupby('platform')

# Determine total sales grouped platform
platform_sales = grouped_by_platform['total_sales'].sum().reset_index()

# Plot bar charts by platform
for platform, group_data in grouped_by_platform:
    group_data.groupby('year_of_release')['total_sales'].sum().plot(kind='bar', title=f"{platform}'s Total Sales per Year", color='green')
    
    plt.xlabel('Year')
    plt.ylabel('Total Sales')
    plt.tight_layout()
    plt.show()
```

## Hyptohesis Testing
### This test is to determine whether or not average user ratings of the Xbox One and PC platforms are the same.


##### Stating  hypthesis 

- μPC = sample mean of user ratings for the PC platform
- μXbox = sample mean of user ratings for the Xbox One platform

##### **Null Hypothesis**
H0: μXbox = μPC 
Sample mean of user ratings of Xbox is equal to sample mean of PC platform.

##### **Alternate Hypothesis:** 
H1: μXbox ≠ μPC
Sample mean of user ratings of Xbox is not equal to sample mean of PC platform.

##### **Alpha Value:** 
- 0.05 (5%)
A significance level of 5% was selected for testing these assumptions.


```python
# collecting samples for PC and XOne game plateform
pc_sample = platform_sales_filtered_13_16[platform_sales_filtered_13_16['platform'] == "PC"]['user_score'].dropna().to_list()
x_one_sample = platform_sales_filtered_13_16[platform_sales_filtered_13_16['platform'] == "XOne"]['user_score'].dropna().to_list()

# Setting up the alpha level
alpha = 0.05 
pvalue1 = st.ttest_ind(pc_sample, x_one_sample) # unpack results

print(f"The pvalue: {pvalue1}")
if pvalue1.pvalue < alpha:
    print(f"We reject the null hypothesis at a {alpha} significance level")
else:
    print(f"We can't reject the null hypothesis at a {alpha} significance level")
```

- **Conclusion**
At a 5% significance level, there is not enough evidence to reject the null hypothesis. Therefore, we conclude that there is no significant difference between the sample mean user ratings for Xbox One and PC platforms.

This means the observed difference in user ratings between Xbox One and PC platforms is likely due to random variation in the sample data, rather than a true difference in the population means.

### This test is to determine whether or not average user ratings of Action and Sports genre are the same.


```python
# collecting samples
action_sample = platform_sales_filtered_13_16[platform_sales_filtered_13_16['genre'] == "Action"]['user_score'].dropna().to_list()
sports_sample = platform_sales_filtered_13_16[platform_sales_filtered_13_16['genre'] == "Sports"]['user_score'].dropna().to_list()

alpha = 0.05
results_action_sports = st.ttest_ind(action_sample, sports_sample)
pvalue2 = results_action_sports

print(f"The pvalue: {pvalue2}")
if pvalue2.pvalue < alpha:
    print(f"We reject the null hypothesis at a {alpha} significance level")
else:
    print(f"We can't reject the null hypothesis at a {alpha} significance level")
```

**Conclusion:** There is a statistically significant difference in the mean user scores between the "Action" and "Sports" genres during the filtered period (2013–2016).
